```python
import pandas as pd

# Load data from CSV files
investments_df = pd.read_csv('data/investments.csv')
applications_df = pd.read_csv('data/applications.csv')
returns_df = pd.read_csv('data/returns.csv')
```

```python
# Initialize an empty DataFrame for the profit history log
profit_history_df = pd.DataFrame(columns=['date', 'investment_name', 'total_return', 'applications_after_last_return', 'profit'])

print(applications_df)
print(returns_df)
print(profit_history_df)
```

```python
# Calculate profit history for each investment
for index, return_row in returns_df.iterrows():
    investment_name = return_row['investment_name']
    date = return_row['date_received']
    total_return = return_row['total_amount']

    last_return = returns_df[(returns_df['investment_name'] == investment_name) & (returns_df['date_received'] < date)].sort_values(by=['date_received'], ascending=False)


    # Calculate the profit percentage
    if last_return.empty:
        last_return_value = investments_df[(investments_df['investment_name'] == investment_name)].iloc[0]['investment_value']
        last_return_date = investments_df[(investments_df['investment_name'] == investment_name)].iloc[0]['investment_date']
    else:
        last_return_value = last_return.iloc[0]['total_amount']
        last_return_date = last_return.iloc[0]['date_received']

    # Filter applications made after the last return date
    filtered_applications = applications_df[(applications_df['investment_name'] == investment_name) & (applications_df['application_date'] > last_return_date) & (applications_df['application_date'] < date)]

    applications_after_last_return = filtered_applications['application_value'].sum()

    profit_percentage = ((total_return + applications_after_last_return - last_return_value) / last_return_value) * 100

    # Append to profit history log
    profit_history_df = pd.concat([pd.DataFrame({
        'date': [date],
        'investment_name': [investment_name],
        'total_return': [total_return],
        'last_return_value': [last_return_value],
        'applications_after_last_return': [applications_after_last_return],
        'profit_percentage': [profit_percentage]
    }), profit_history_df], ignore_index=True)
```
